## Datetime Format 2019-2023 vereinheitlichen

```python
from datetime import datetime
```

```python
df2 = pd.read_csv('/Users/leabeckmann/Library/CloudStorage/OneDrive-PersoÌˆnlich/Alle Dokumente/UNI/ADS02 - ML/Daten/combined_data.csv')
```

```python
df2.head()
```

```python
#Dictionary fÃ¼r die Umbenennung, weil die Spalten in d1 und d2 anders benannt sind
rename_dict = {
    'nbhd': 'Nbhd',
    'Rooms': 'Nr_of_rms',
    'FinishedSqft': 'Fin_sqft'
}

# Benenne die Spalten um
df2.rename(columns=rename_dict, inplace=True)

# Speichere die angepasste DataFrame in einer neuen CSV-Datei
df2.to_csv('/Users/leabeckmann/Library/CloudStorage/OneDrive-PersoÌˆnlich/Alle Dokumente/UNI/ADS02 - ML/Daten/combined_data_mod.csv', index=False)
```

```python
df2.head()
```

```python
unique_dates = df2['Sale_date'].unique()
print(unique_dates[:10]) 
```

```python
#Hier war die Herausforderung, dass teilweise verschiedene Datumsformate angegeben waren
#und es nicht vereinheitlicht werden konnte
def convert_date(date_str):
    if pd.isnull(date_str) or date_str in ["nan", "NaT", "None"]:  
        return None
    for fmt in ('%m/%d/%Y', '%Y-%m-%d'):
        try:
            return datetime.strptime(date_str, fmt).strftime('%Y-%m-%d')
        except ValueError:
            pass
    raise ValueError(f"No valid date format found for {date_str}")

#Alle Werte konvertieren
df2['Sale_date'] = df2['Sale_date'].astype(str)

df2['Sale_date'] = df2['Sale_date'].apply(convert_date)

#Konvertieren in Datetime
df2['Sale_date'] = pd.to_datetime(df2['Sale_date'], errors='coerce')

#Konvertieren in Jahr-Monat (wie in 2002-2018 Datei)
df2['Sale_date'] = df2['Sale_date'].dt.strftime('%Y-%m')

print(df2.head())
```

```python
# Dtypes konvertierung

df2['District'] = df2['District'].astype('int64')
df2['Nbhd'] = df2['Nbhd'].astype('int64')
df2['Year_Built'] = df2['Year_Built'].astype('int64')
df2['Nr_of_rms'] = df2['Nr_of_rms'].astype('int64')

# Kommata entfernen und konvertieren
df2['Fin_sqft'] = df2['Fin_sqft'].astype(str).str.replace(',', '').astype(float)
df2['Fin_sqft'] = df2['Fin_sqft'].round().astype('int64')

df2['Units'] = df2['Units'].astype('int64')
df2['Bdrms'] = df2['Bdrms'].astype('int64')
df2['Fbath'] = df2['Fbath'].astype('int64')
df2['Hbath'] = df2['Hbath'].astype('int64')
df2['Lotsize'] = df2['Lotsize'].str.replace(',', '').astype(float).round().astype('int64')

# Nicht numerische Werte replacen und konvertieren
df2['Sale_price'] = df2['Sale_price'].str.replace('[^\d]', '', regex=True).astype('int64')
```
